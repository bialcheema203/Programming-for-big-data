# Question#1 Find the total number of patients of each disease (diagnosis) and plot in histogram.

```python
import pandas as pd
import matplotlib as plt
from matplotlib import pyplot
```

```python
md = pd.read_csv('jsrt_metadata.csv')
print(md['diagnosis'].value_counts())
```

```python
md['diagnosis'] = md['diagnosis'].astype('str')
plt.pyplot.hist(md['diagnosis'],color='blue',ec='lightgreen',orientation='horizontal')
plt.pyplot.title('Number of People Diagnosis')
plt.pyplot.xlabel('Frequency')
plt.pyplot.ylabel('Category')
```

# Question#2 . Find the total number of patients of each disease (diagnosis) with respect to the disease state(malignant or benign) and plot in a pie chart.

```python
print(md[['diagnosis','state']].value_counts())
```

```python
plt.pyplot.pie(md[['diagnosis','state']].value_counts(),shadow=True,autopct='%1.1f%%')
```
